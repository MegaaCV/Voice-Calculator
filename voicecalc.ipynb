```python
import tkinter as tk
import speech_recognition as sr
import pyttsx3
import sympy as sp
from word2number import w2n
import re
import math


class VoiceCalculator:
    def __init__(self, root):
        self.root = root
        self.root.title("Voice Recognizable Calculator")
        self.root.geometry("600x420")
        self.root.resizable(False, False)

        self.result_var = tk.StringVar()
        self.result_var.set("Click 'Start' to begin...")

        self.label = tk.Label(root, textvariable=self.result_var,
                              font=("Arial", 14), wraplength=580, justify="center")
        self.label.pack(pady=40)

        self.start_button = tk.Button(root, text="ðŸŽ¤ Start",
                                      command=self.start_calculation, font=("Arial", 14))
        self.start_button.pack(pady=20)

        
        self.engine = pyttsx3.init()
        self.engine.setProperty("rate", 150)

    def speak(self, text):
        self.engine.say(text)
        self.engine.runAndWait()

    def words_to_numbers(self, text: str) -> str:
        """Convert number words (including decimals like 'two point five') to digits."""
        try:
            
            text = text.replace(" point ", ".")
            return str(w2n.word_to_num(text))
        except:
            return text

    def preprocess_expression(self, expr: str) -> str:
        expr = expr.lower().strip()

       
        corrections = {
            "sign of": "sine of",
            "sin of": "sine of",
            "turn off": "tangent of",
            "tan of": "tangent of",
            "ten of": "tangent of",
            "cause of": "cosine of",
            "cos of": "cosine of",
        }
        for wrong, right in corrections.items():
            expr = expr.replace(wrong, right)

        
        expr = expr.replace("multiplied by", "*")
        expr = expr.replace("times", "*")
        expr = expr.replace("x", "*")
        expr = expr.replace("plus", "+")
        expr = expr.replace("minus", "-")
        expr = expr.replace("divide", "/")
        expr = expr.replace("divided by", "/")
        expr = expr.replace("over", "/")
        expr = expr.replace("power", "**")

        
        expr = self.words_to_numbers(expr)

        
        functions = {
            r"square root of (-?\d+(\.\d+)?)": r"sqrt(\1)",
            r"sine of (-?\d+(\.\d+)?)": r"sin(\1*pi/180)",
            r"cosine of (-?\d+(\.\d+)?)": r"cos(\1*pi/180)",
            r"tangent of (-?\d+(\.\d+)?)": r"tan(\1*pi/180)",
            r"logarithm of (-?\d+(\.\d+)?)": r"log(\1,10)",
            r"log of (-?\d+(\.\d+)?)": r"log(\1,10)",
            r"ln of (-?\d+(\.\d+)?)": r"log(\1)",
            r"factorial of (-?\d+(\.\d+)?)": r"factorial(\1)"
        }

        for pattern, repl in functions.items():
            expr = re.sub(pattern, repl, expr)

        return expr.strip()

    def start_calculation(self):
        self.speak("Please say your calculation.")
        self.result_var.set("Listening...")
        self.root.update()

        recognizer = sr.Recognizer()
        with sr.Microphone() as source:
            try:
                audio = recognizer.listen(source, timeout=5, phrase_time_limit=5)
                expression = recognizer.recognize_google(audio)
                self.result_var.set("You said: " + expression)
                self.root.update()

                expr = self.preprocess_expression(expression)
                print("Processed expression:", expr) 

                try:
                    symbolic_expr = sp.sympify(expr)
                    result = sp.N(symbolic_expr)  
                    result = round(result, 6)     
                    self.result_var.set(f"Result: {result}")
                    self.speak(f"The result is {result}")
                except Exception as e:
                    print("Error:", e)
                    self.result_var.set("Sorry, could not calculate.")
                    self.speak("Sorry, I could not calculate that.")

            except sr.UnknownValueError:
                self.result_var.set("Sorry, I didn't understand.")
                self.speak("Sorry, I didn't understand.")
            except sr.RequestError:
                self.result_var.set("Speech service error.")
                self.speak("Speech service error.")
            except sr.WaitTimeoutError:
                self.result_var.set("No speech detected.")
                self.speak("No speech detected.")


if __name__ == "__main__":
    root = tk.Tk()
    app = VoiceCalculator(root)
    root.mainloop()
```

```python
import customtkinter as ctk
import speech_recognition as sr
import sympy as sp
import pyttsx3
from threading import Thread


# Init Speech

recognizer = sr.Recognizer()
engine = pyttsx3.init()
running = False   # Flag to control voice loop

def speak(text):
    print("ðŸ”Š", text)
    engine.say(text)
    engine.runAndWait()


# Preprocessor

def preprocess(expression: str) -> str:
    expr = expression.lower().strip()

    # --- Fix common mishears ---
    replacements = {
        "sign of": "sine of",
        "sin of": "sine of",
        "sine off": "sine of",

        "cause of": "cosine of",
        "cos of": "cosine of",
        "coss of": "cosine of",
        "cosign of": "cosine of",

        "turn off": "tangent of",
        "tan of": "tangent of",
        "ten of": "tangent of",
    }
    for wrong, correct in replacements.items():
        expr = expr.replace(wrong, correct)

    # --- Operators ---
    expr = expr.replace("plus", "+")
    expr = expr.replace("minus", "-")
    expr = expr.replace("times", "*").replace("multiplied by", "*")
    expr = expr.replace("divided by", "/").replace("over", "/")
    expr = expr.replace("power of", "**").replace("to the power of", "**")

    # --- Brackets ---
    expr = expr.replace("open bracket", "(")
    expr = expr.replace("close bracket", ")")

    # --- Exponents ---
    expr = expr.replace("squared", "**2")
    expr = expr.replace("cubed", "**3")

    # --- Functions ---
    if "square root of" in expr:
        expr = expr.replace("square root of", "sqrt(") + ")"

    if "log of" in expr:
        expr = expr.replace("log of", "log(") + ")"
    if "ln of" in expr:
        expr = expr.replace("ln of", "ln(") + ")"

    if "sine of" in expr:
        expr = expr.replace("sine of", "sin(pi/180*") + ")"
    if "cosine of" in expr:
        expr = expr.replace("cosine of", "cos(pi/180*") + ")"
    if "tangent of" in expr:
        expr = expr.replace("tangent of", "tan(pi/180*") + ")"

    return expr

# Calculation

def calculate(expression: str):
    try:
        processed = preprocess(expression)
        print(f"[DEBUG] Heard: {expression} | Processed: {processed}")
        expr = sp.sympify(processed, evaluate=True)

        # Force evaluation of trig and roots
        result = sp.N(expr, 15)  

        # Always return numeric float when possible
        if result.is_real:
            return float(result)
        else:
            return result
    except Exception as e:
        return f"Error: {e}"


# Voice Calculator
def voice_calculator():
    global running
    with sr.Microphone() as source:
        recognizer.adjust_for_ambient_noise(source, duration=1)
        speak("Voice calculator is ready. Say quit to exit.")

        while running:
            try:
                audio = recognizer.listen(source, timeout=5, phrase_time_limit=7)
                if not running:
                    break
                text = recognizer.recognize_google(audio)

                if text.lower() in ["quit", "exit", "stop"]:
                    speak("Goodbye!")
                    update_gui("Goodbye!", "")
                    stop_app()
                    break

                result = calculate(text)
                update_gui(text, result)
                add_to_history(text, result)
                speak(f"The result is {result}")

            except sr.WaitTimeoutError:
                continue
            except sr.UnknownValueError:
                if running:
                    update_gui("Could not understand", "")
                    speak("I did not understand, please repeat")
            except Exception as e:
                if running:
                    update_gui("Error", e)
                    speak("Sorry, I could not calculate that.")


def update_gui(expression, result):
    expression_var.set(f"Expression: {expression}")
    result_var.set(f"Result: {result}")

def start_voice_thread():
    global running
    if not running:
        running = True
        t = Thread(target=voice_calculator, daemon=True)
        t.start()

def calculate_from_text():
    expr = text_entry.get()
    if expr.strip():
        result = calculate(expr)
        update_gui(expr, result)
        add_to_history(expr, result)
        speak(f"The result is {result}")

def add_to_history(expression, result):
    history_box.insert("end", f"{expression} = {result}\n")
    history_box.see("end")

def stop_app():
    global running
    running = False
    root.after(100, root.destroy)  # delay to allow thread to stop


# GUI Setup

ctk.set_appearance_mode("dark")
ctk.set_default_color_theme("blue")

root = ctk.CTk()
root.title("ðŸŽ¤ Smart Voice + Text Calculator")
root.geometry("700x500")

expression_var = ctk.StringVar()
result_var = ctk.StringVar()

title_label = ctk.CTkLabel(root, text="ðŸŽ¤ Smart Calculator", font=("Arial Rounded MT Bold", 24))
title_label.pack(pady=10)

expr_label = ctk.CTkLabel(root, textvariable=expression_var, font=("Arial", 16))
expr_label.pack(pady=10)

res_label = ctk.CTkLabel(root, textvariable=result_var, font=("Arial", 18))
res_label.pack(pady=10)

# Text Input
input_frame = ctk.CTkFrame(root)
input_frame.pack(pady=10)

text_entry = ctk.CTkEntry(input_frame, width=300, font=("Arial", 14))
text_entry.grid(row=0, column=0, padx=10)

calc_button = ctk.CTkButton(input_frame, text="Calculate", command=calculate_from_text)
calc_button.grid(row=0, column=1, padx=5)

# Buttons
btn_frame = ctk.CTkFrame(root)
btn_frame.pack(pady=10)

voice_btn = ctk.CTkButton(btn_frame, text="ðŸŽ¤ Start Voice Mode", command=start_voice_thread, width=200)
voice_btn.grid(row=0, column=0, padx=10)

exit_btn = ctk.CTkButton(btn_frame, text="Exit", command=stop_app, fg_color="red", width=100)
exit_btn.grid(row=0, column=1, padx=10)

# History Box
history_label = ctk.CTkLabel(root, text="ðŸ“œ History", font=("Arial", 16))
history_label.pack(pady=5)

history_box = ctk.CTkTextbox(root, width=600, height=150, font=("Consolas", 12))
history_box.pack(pady=5)

# Handle Close
root.protocol("WM_DELETE_WINDOW", stop_app)

# Run GUI
root.mainloop()
```
